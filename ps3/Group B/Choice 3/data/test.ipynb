```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('data.csv')
df['agreement'] = df['Agreement (0=complete disagreement; 5=complete agreement)'] 
df['pid'] = df['participant'].astype(str) + '_' + df['group'].astype(str)
df.head()
```

```python
for p, group in df.groupby(['pid']):
    print(len(group))
```

```python
output = {}
for (participant, scale), group in df.groupby(['pid', 'subscale']):
    culture = group['group'].iloc[0]
    if culture not in output:
        output[culture] = {}
    if scale not in output[culture]:
        output[culture][scale] = []
    output[culture][scale].append(np.mean(group['agreement'].values))


```

```python
stats.ttest_ind(output['Moroccan']['PAST'], output['young Spaniard']['PAST'])
```

```python
stats.ttest_ind(output['Moroccan']['FUTURE'], output['young Spaniard']['FUTURE'])
```
